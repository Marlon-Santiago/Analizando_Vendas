# Missão 1: Time de vendas

Precisamos construir uma visualização para indicar os **10 maiores clientes por total de venda**.

Com isso, a equipe de vendas pode pensar em estratégias para manter esses clientes comprando em nossa loja.

```python
dados = pd.read_csv('relatorio_vendas.csv')
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados.data_pedido = pd.to_datetime(dados.data_pedido)
dados.info()
```

```python
top_10_vendas = dados.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
top_10_vendas = top_10_vendas.reset_index()
top_10_vendas
```

```python
top_10_vendas.columns = ['Clientes','Vendas']
top_10_vendas['Ranque'] = top_10_vendas.index + 1
top_10_vendas.set_index('Ranque', inplace = True)
top_10_vendas
```

```python
s = top_10_vendas.style
s
```

```python
s.format({'Vendas': 'R${:,.2f}'})
s
```

# Missão 2: Time comercial

Precisamos construir uma visualização que permita associar o **total ganho em vendas e o total lucrado por tipo de produto**.

Isso vai permitir que o time comercial possa focar em estratégias para alavancar as rendas que ganham com cada tipo de produto vendido na loja.

```python
lucro_por_produto = dados.groupby(['tipo_produto'])[['vendas','lucro']].sum()
lucro_por_produto.reset_index()
lucro_por_produto.index.name = 'tipo produto'
lucro_por_produto
```

```python
stilo_produto = lucro_por_produto.style
stilo_produto
```

```python
stilo_produto.format('R$ {:,.2f}')\
                     .highlight_max(color = 'blue')\
                     .highlight_min(color = 'red')

stilo_produto
```

```python
stilo_produto.format('R$ {:,.2f}').background_gradient(cmap = 'Blues')
stilo_produto
```

```python
cabecalho = {
    'selector':'th',
    'props':'font-weight: bold; font-family: Arial; text-aling: center; text-transform: capitalize;'
}
stilo_produto.set_table_styles([cabecalho], overwrite = False)
```

# Missão 3: Time de logística

Precisamos construir uma visualização que permita entender a **distribuição de pedidos por região do Brasil**.

Isso vai permitir que o time logística entenda onde é mais interessante focar material e pessoas.

```python
dados.head()
```

```python
dados_por_regiao = pd.DataFrame(dados.regiao.value_counts())
dados_por_regiao['N° Pedidos'] = dados_por_regiao
dados_por_regiao
```

```python
import numpy as np
```

```python
dados_por_regiao.index.name = 'Região'
```

```python
porcentagem = dados_por_regiao['N° Pedidos'].to_numpy()
porcentagem = 100 * porcentagem / porcentagem.sum()
porcentagem
```

```python
dados_por_regiao['Porcentagem'] = porcentagem
dados_por_regiao
```

```python
estilo_regiao = dados_por_regiao.style
estilo_regiao
```

```python
cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; color : #EE82EE; text-align: right; background-color: black'
}
celulas = {
    'selector': 'td',
    'props': 'font-weight: bold; font-family: Arial; color : #FF69B4; text-align: right; background-color: black;'
}

estilo_regiao.set_table_styles([cabecalho,celulas])
```

```python
estilo_regiao.format({'Porcentagem' : '{:.2f} %'})\
             .bar(subset = 'Porcentagem', vmin = 0, vmax = 100.0, color = 'white')
```

# Missão 4: Time de logística

Novamente o time de logística deseja repensar sua estrutura e organização de fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
dados.head()
```

```python
dados_meses = dados.copy()
dados_meses.sort_values('data_pedido')
dados_meses
```

```python
dados_meses['Meses'] = dados_meses['data_pedido'].dt.strftime('%Y - %b')
dados_meses = dados_meses.reset_index(drop = True)
dados_meses
```

```python
vendas_mensais = dados_meses.pivot_table(index = 'departamento', columns = 'Meses', values = 'quantidade',
                                         aggfunc = 'sum', sort = False)
vendas_mensais
```

```python
estilo_mensal = vendas_mensais.style
estilo_mensal
```

```python
estilo_mensal.set_sticky(axis = 'index')
```

```python
colunas = {
    'selector' : '.col_heading',
    'props' : 'font-weight : normal; font-family : Arial'
}
tabelas = {
    'selector' : 'td,th',
    'props' : 'text-align : left; color : red'
}
index = {
    'selector' : '.index_name',
    'props': 'font-weight : bold; font-family : Arial; color : red; text-align : right;'
}
estilo_mensal.set_table_styles([colunas,tabelas,index],overwrite = False)
```

# Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa. Nele, todos que trabalham na loja podem entender como anda a empresa. Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos construir uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**. Com isso, será possível destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geraram ganho em vendas.

```python
dados.head()
```

```python
dados_cliente = dados.pivot_table(index = 'segmento_cliente', columns = 'modo_envio',values = 'vendas',
                                  aggfunc = 'sum')
dados_cliente
```

```python
dados_cliente['Total'] = dados_cliente.sum(axis = 1)
dados_cliente.loc['Total'] = dados_cliente.sum()
dados_cliente
```

```python
compra_cliente = dados_cliente.style.format('R$ {:,.2f}')
compra_cliente
```

```python
tabela = {
    'selector' : 'td,th:not(.index_name)', 
    'props' : 'font-weight : normal; font-family : Arial; text-align : center; background-color : black'
}
index = {
    'selector' : '.index_name',
    'props' : 'font-weight : normal; text-align : rigth; font-style : italic; color : cinza'
}
compra_cliente.set_table_styles([tabela,index])
```

```python
compra_cliente.set_table_styles({
    'Total': [{ 
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    { 
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }],
    'B2B': [{ 
        'selector': 'th',
        'props': 'border-top: 1px solid #181818 '
    },
    { 
        'selector': 'td',
        'props': 'border-top: 1px solid #181818 '
    }]
}, overwrite = False, axis = 1)
```

```python
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props': 'background-color: #181818'
    }]
}, overwrite=False, axis=0)

cores_coluna = pd.DataFrame(['false', 'true', 'false'], index= dados_cliente['Total'].index,
                             columns = ['Total'])

compra_cliente.set_td_classes(cores_coluna)
```
